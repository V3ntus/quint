```python
def get_transcripts(links):
    transcripts = []
    for each in tqdm(links):
        try:
            page = requests.get(each).text
            soup = BeautifulSoup(page)
            mydivs = soup.find_all("section", {"class": "transcripts"})

            transcript = []
            for each in mydivs[0].find_all("div", {"class": "transcript is-main"}):
                text = each.find("p", {"class": "transcript__text tg-paragraph"}).text.replace('\n', '')
                speaker = each.find("li", {"class": "speaker"}).text
                timestamp = each.find("li", {"class": "tg-caption"}).text
                transcript.append({'timestamp':timestamp,
                                   'speaker':speaker,
                                   'text':text,
                                  })
            transcripts.append(transcript)
            #print('Succcess')
        except Exception as inst:
            pass
            #print(f'Error {inst}')
    return transcripts
get_transcripts(links[:4])
```

```python
import requests
from bs4 import BeautifulSoup
def get_links_on_page(num):
    params = {'page':num, 'search':'speaker', 'category':'transcripts'}
    url = 'https://steno.ai/search'
    response = requests.get(url=url, params=params).text
    soup = BeautifulSoup(response)
    page = soup.find_all('a', {'class':'article__episode tg-subtitle'}, href=True)#[1]['href']
    temp_links = []
    for each in page:
        temp_links.append(each['href'])
    return temp_links
```
